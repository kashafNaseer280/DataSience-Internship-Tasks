```python
import pandas as pd   #libray used to analyze and mainupulate data
mydataset = pd.read_csv("C:\\Users\\dell\\Downloads\\iris.csv") #loading iris dataset
print(mydataset.shape) #showing how many rows and columns in dataset
print(mydataset.columns) # will display names of the columns
print(mydataset.head()) #will display first five rows of dataset
```

```python
import matplotlib.pyplot as plt #library used for visualization
plt.scatter(mydataset['petal_length'], mydataset['petal_width'], color='red') #realationship between the columns 
plt.xlabel('petal length') #just giving name to x-axis in order to remember what y-axis is showing
plt.ylabel('Petal width') #iving name to y-axis in order to remember what y-axis is showing
plt.show() # show scatter plot
```

```python
plt.hist(mydataset['sepal_width'],bins=10, color='orange') #it will make histogram of sepal_width , will divide sepal_width into 10 equal parts, will show the bar color as orange
plt.xlabel('Sepal Width')
plt.ylabel('Flower Count')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(mydataset['sepal_width'], color='orange')
plt.title('Box blot')
plt.show()
```
